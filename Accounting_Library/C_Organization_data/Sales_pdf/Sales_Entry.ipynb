```python
def enter_leger(C5):
    

    main_df = pd.read_excel('main_df.xlsx',engine='openpyxl')
    if C5.lower() in list(main_df['C5']) :
       print(f"Ledger Name {C5} already present")
       c1   = main_df.loc[main_df['C5'] == C5,'C1'].iloc[0]
       c2   = main_df.loc[main_df['C5'] == C5,'C2'].iloc[0]
       c3   = main_df.loc[main_df['C5'] == C5,'C3'].iloc[0]
       return c1,c2,c3

    C3 = input("Enter C3 (Account Name) ")
    C4 = input("Enter C4 (Sub ledger) ")
    
    if (main_df.loc[main_df['C3'] == C3, 'C5'] == '').any(): 
        if C4.lower() =='none':
            main_df.loc[main_df['C3'] == C3,'C4'] =  C3

        else:
            main_df.loc[main_df['C3'] == C3,'C4'] =  C4

        main_df.loc[main_df['C3'] == C3,'C5'] =  C5.lower()
        c1   = main_df.loc[main_df['C3'] == C3,'C1'].iloc[0]
        c2   = main_df.loc[main_df['C3'] == C3,'C2'].iloc[0]
        c3   = main_df.loc[main_df['C3'] == C3,'C3'].iloc[0]
        c4   = main_df.loc[main_df['C3'] == C3,'C4'].iloc[0]
        c5   = main_df.loc[main_df['C3'] == C3,'C5'].iloc[0]
        main_df.to_excel('main_df.xlsx')
        return c1,c2,c3

    else:
        if C4.lower() =='none':
            C4 =  C3
            
        c1 = main_df[main_df['C3']==C3]['C1'].iloc[0]
        c2 = main_df[main_df['C3']==C3]['C2'].iloc[0]
        c3 = main_df[main_df['C3']==C3]['C3'].iloc[0]
        g_code = main_df[main_df['C3']==C3]['G_CODE'].iloc[0]

        new_data = {'C1': c1, 'C2': c2, 'C3':c3 ,'C4':C4,'C5': C5.lower() , 'G_CODE':g_code}
        main_df = main_df.append(new_data, ignore_index=True)
        main_df.to_excel('main_df.xlsx')
        return c1,c2,c3

```

```python
df = pd.read_excel('s3.xlsx',engine='openpyxl')
```

```python
from dr_cr_rule import ASSET_Expense_Accounting
```

```python
for i in range(len(df)):
    en = df.iloc[i]
    tx = en['Tax']
    sys = ASSET_Expense_Accounting()
    
    date = en['Invoice Date']
    invoice_number = en['Invoice Number']
    journal_number = 'S00'+str(i)
    
    def To_Entry():
        l1 = en['To']
        print(l1)
        if l1.lower() not  in list(main_df['C5']) :
            c1,c2,c3 = enter_leger(l1.lower())
    
        if c1 == 'ASSET' or c1=='EXPENSE':
            dr = sys.debit_transaction(en['Total'])
            cr = 0
    
            k['J8 - Closing Balance'].append(dr)
            k['J5 - Opening Balance'].append(0)
            k['J9 - Description/Narration'].append('Sales')
            k['J3 - Invoice_Ref_Number'].append(invoice_number)
            k['J4 - Ledger_Name'].append(l1)
            k['J2 - Date'].append(date)
            k['J1 - Journal Number'].append(journal_number)
            
            k['J6 - Dr Amount'].append(dr)
            k['J7 - Cr Amount'].append(cr)
    
    def tax_igst_entry():
        igst_c5 = 'tax igst' + ' ' + str(tx) + '%'
        print(igst_c5)
    
        if igst_c5.lower() not  in list(main_df['C5']) :
            c1,c2,c3 = enter_leger(igst_c5.lower())
    
        if c1 == 'ASSET' or c1=='EXPENSE':
            amu = en['Tax IGST']
            cr = sys.credit_transaction(amu)
            print(cr)
            dr = 0
    
            k['J8 - Closing Balance'].append(cr)
            k['J5 - Opening Balance'].append(0)
            k['J9 - Description/Narration'].append('Sales')
            k['J3 - Invoice_Ref_Number'].append(invoice_number)
            k['J4 - Ledger_Name'].append(igst_c5)
            k['J2 - Date'].append(date)
            k['J1 - Journal Number'].append(journal_number)
            
            k['J7 - Cr Amount'].append(cr)
            k['J6 - Dr Amount'].append(dr)
    
    
    def tax_cgst_entry():
        cgst_c5 = 'tax cgst' + ' ' + str(tx) + '%'
        print(cgst_c5)
    
        if cgst_c5.lower() not  in list(main_df['C5']) :
            c1,c2,c3 = enter_leger(cgst_c5.lower())
    
        if c1 == 'ASSET' or c1=='EXPENSE':
            tc = en['Tax CGST']
            cr = sys.credit_transaction(tc)
            dr = 0
            k['J8 - Closing Balance'].append(cr)
            k['J5 - Opening Balance'].append(0)
            k['J9 - Description/Narration'].append('Sales')
            k['J3 - Invoice_Ref_Number'].append(invoice_number)
            k['J4 - Ledger_Name'].append(cgst_c5)
            k['J2 - Date'].append(date)
            k['J1 - Journal Number'].append(journal_number)
            
            k['J7 - Cr Amount'].append(cr)
            k['J6 - Dr Amount'].append(dr)
    
    def To_Sales():
        sales = 'To Sales'
        print(sales)
        if sales.lower() not  in list(main_df['C5']) :
            c1,c2,c3 = enter_leger(sales.lower())
    
        if c1 == 'ASSET' or c1=='EXPENSE':
            amu = en['Amount']
            cr = sys.credit_transaction(amu)
            dr = 0
            
            k['J8 - Closing Balance'].append(cr)
            k['J5 - Opening Balance'].append(0)
            k['J9 - Description/Narration'].append('Sales')
            k['J3 - Invoice_Ref_Number'].append(invoice_number)
            k['J4 - Ledger_Name'].append(sales)
            k['J2 - Date'].append(date)
            k['J1 - Journal Number'].append(journal_number)
            
            k['J7 - Cr Amount'].append(cr)
            k['J6 - Dr Amount'].append(dr)
    
    
    To_Entry()
    tax_igst_entry()
    tax_cgst_entry()
    To_Sales()
```
